Создание объектов в NumPy

```python
series_int=pd.Series(np.arange(5, 27)) # создать следовательность целых чисел
print(series_int)

series_linspace=pd.Series(np.linspace(12,26,5)) # Создать посл-ть из 5 чисел между 12 и 26
print(series_linspace)
```

```python
np.random.seed(12345) # Генератор случайных чисел

normal=pd.Series(np.random.normal(size=1000)) # Создать Series из 5 нормально распределённых случайных чисел
normal.plot()
```
